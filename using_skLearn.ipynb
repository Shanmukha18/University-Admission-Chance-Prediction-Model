***Multiple Linear Regression Model Using scikit-learn***

Importing the necessary Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Loading the Data Set

```python
df = pd.read_csv("DataSet.csv")
```

Defining Features and Response
- Separating all the features as x and responses as y.
- So the columns are features(x) and column is responce(y).

```python
x = df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']].values
y = df['Chance of Admit '].values
```

Separting training and testing data
- Separate the dataset into training and testing
- We take out certain proportion of data for testing purpose so that we calculate the models accuracy later.
- Over here we take out 20% of data for testing purpose.

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split( x, y, test_size = 0.80,random_state=0)
```

Checking the shape of train and test data

```python
Ytrain.shape, Ytest.shape
```

Importing the LinearRegression from scikit-learn
- sklearn has inbuilt class called LinearRegression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Training the model
- Training the model using fit function from LinearRegression class
- by doing so we can obtain intercept and weights

```python
model.fit(Xtrain, Ytrain)
```

Knowing the weights and bias
- by using intercept_ function we can obtain the value bias
- by using coeff_ function we can obtain the values of weights

```python
print("Intercept:\ntheta 0 = ", model.intercept_)
print("\nX Coefficients: ")
for i in range(len(model.coef_)):
    print("theta",i+1,"= ",model.coef_[i])
```

Predicting the Test Data Set

```python
y_pred = model.predict(Xtest)
```

Calculating Metrics
- Calculate Mean Absolute Error and Mean Squarred Error using metrics module
- Evaluate the accuracy of our model using score function

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
print("Mean Absolute Error: ",mean_absolute_error(Ytest, y_pred))
print("Mean Squared Error: ",mean_squared_error(Ytest, y_pred))
print("Accuracy Percentage: ", round((model.score(Xtrain, Ytrain)*100),2), "%")
```

Plotting the bestfit line in the graph

```python
plt.scatter(Ytest, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Values")
plt.plot([min(Ytest), max(Ytest)], [min(Ytest), max(Ytest)], color='red')
plt.show()
```
